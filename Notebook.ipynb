```python
import numpy as np
import pandas as pd
```

```python
mouse_train = pd.read_csv('mouse_data_enakano_train.csv')
mouse_train.head()
```

```python
def generate_features(mouse_raw):
  
  horiz_spd = mouse_raw['x'].diff()/mouse_raw['client timestamp'].diff()
  vert_spd = mouse_raw['y'].diff()/mouse_raw['client timestamp'].diff()
  tang_spd = np.sqrt((horiz_spd**2)+(vert_spd**2))
  horiz_spd_ag_dist = (mouse_raw['x'] - mouse_raw['x'][0])/(mouse_raw['client timestamp']-mouse_raw['client timestamp'][0])
  vert_spd_ag_dist = (mouse_raw['y'] - mouse_raw['y'][0])/(mouse_raw['client timestamp']-mouse_raw['client timestamp'][0])
  avrge_spd_ag_dist = np.sqrt((mouse_raw['x'].diff()**2)+(mouse_raw['y'].diff()**2)).sum()/(mouse_raw['client timestamp']-mouse_raw['client timestamp'][0])
 
  return {'horiz_spd':horiz_spd, 'vert_spd':vert_spd, 'tang_spd':tang_spd,'horiz_spd_ag_dist':horiz_spd_ag_dist, 'vert_spd_ag_dist':vert_spd_ag_dist, 'avrge_spd_ag_dist':avrge_spd_ag_dist}
        
def generate_features_2(mouse_raw, mouse_raw_1):  
  
  horiz_acc = mouse_raw_1['horiz_spd'].diff()/mouse_raw['client timestamp'].diff()
  vert_acc = mouse_raw_1['vert_spd'].diff()/mouse_raw['client timestamp'].diff()
  tang_acc = np.sqrt((horiz_acc**2)+(vert_acc**2))
  horiz_acc_ag_dist = (mouse_raw_1['horiz_spd_ag_dist'] - mouse_raw_1['horiz_spd_ag_dist'][0])/(mouse_raw['client timestamp']-mouse_raw['client timestamp'][0])
  vert_acc_ag_dist= (mouse_raw_1['vert_spd_ag_dist'] - mouse_raw_1['vert_spd_ag_dist'][0])/(mouse_raw['client timestamp']-mouse_raw['client timestamp'][0])
  avrge_acc_ag_dist = (mouse_raw_1['avrge_spd_ag_dist'] - mouse_raw_1['avrge_spd_ag_dist'][0])/(mouse_raw['client timestamp']-mouse_raw['client timestamp'][0])
 
  return {'horiz_acc':horiz_acc, 'vert_acc':vert_acc, 'tang_acc':tang_acc, 'horiz_acc_ag_dist':horiz_acc_ag_dist, 'vert_acc_ag_dist':vert_acc_ag_dist, 'avrge_acc_ag_dist':avrge_acc_ag_dist}
```

```python
mouse_train_1 = pd.DataFrame(generate_features(mouse_train)).fillna(0)

from numpy import inf
mouse_train_1[mouse_train_1 == inf] = 0
mouse_train_1.head()
```

```python
mouse_train_2 = pd.DataFrame(generate_features_2(mouse_train, mouse_train_1)).fillna(0)
mouse_train_2.head()
```

```python
mouse_train_feat = pd.concat([mouse_train_1,mouse_train_2],axis=1)                           
mouse_train_feat.head()
```

Apenas para finalidade de arquivamento das informaÃ§Ãµes, optamos por gravar as "features" em um arquivo .csv

```python
mouse_train_feat.to_csv('mouse_features.csv')
```

AlÃ©m dos dados para o treinamento do modelo, geramos tambÃ©m um conjunto de dados para teste, com o mesmo programa. No caso dos dados abaixo, os tempo de coleta foi de aproximadamente 15 minutos.

```python
mouse_test = pd.read_csv('mouse_data_enakano_test.csv')
mouse_test.head()
```

Novamente, utilizamos a forma de extraÃ§Ã£o de features, acima indicada, para extraÃ§Ã£o das mesmas em relaÃ§Ã£o aos dados de teste.

```python
mouse_test_1 = pd.DataFrame(generate_features(mouse_test)).fillna(0)

mouse_test_1[mouse_test_1 == inf] = 0
mouse_test_1.head()
```

```python
mouse_test_2 = pd.DataFrame(generate_features_2(mouse_test, mouse_test_1)).fillna(0)
mouse_test_2.head()
```

```python
mouse_test_feat = pd.concat([mouse_test_1,mouse_test_2],axis=1)                           
mouse_test_feat.head()
```

Apenas para finalidade de arquivamento das informaÃ§Ãµes, optamos por gravar as "features" em um arquivo .csv

```python
mouse_test_feat.to_csv('mouse_features_test.csv')
```

Como as features forma geradas em dataframes do pandas, convertemos em arrays simples excluindo as colunas referentes aos dados captados inicialmente, assim como a linha dos tÃ­tulo de cada coluna, e a primeira linha de dados, que continha zeros.


O modelo ora utilizado Ã© a classificaÃ§Ã£o dos dados por meio de Suport Vector Machine. No presente caso foi utilizado o scikit-learn.

```python
from sklearn import svm
```

Levando em consideraÃ§Ã£o que estamos utilizando os dados de apenas um usuÃ¡rio, foi utilizado o OneClassSVM, conforme parametros estabelecidos no paper.

Conforme o caso, o ajuste dos parametros do modelo, deverÃ¡ ser ajustado para se determinar a melhor forma do mesmo. (e.g. nu=?, kernel=?, etc.)

```python
clf = svm.OneClassSVM(nu=0.07, kernel="linear")
clf.fit(mouse_train_feat)
```

```python
y_pred_train = clf.predict(mouse_train_feat)
y_pred_test = clf.predict(mouse_test_feat)

n_error_train = y_pred_train[y_pred_train == -1].size
n_error_test = y_pred_test[y_pred_test == -1].size
```
